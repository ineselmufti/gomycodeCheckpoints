```python
def Factorial(n):
    a=1
    for i in range(n,1,-1):
        a=a*i
    return a
Factorial(7)
```

```python
def Dictionary(n):
    dict= {}
    for i in range(1,n+1):
        dict[i]=i*i
    return dict
Dictionary(8)
```
